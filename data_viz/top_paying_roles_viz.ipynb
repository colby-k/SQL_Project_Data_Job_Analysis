```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = {
    "job_id": [226942, 547382, 552322, 99305, 1021647, 168310, 731368, 310660, 1749593, 387860],
    "job_title": [
        "Data Analyst", "Director of Analytics", "Associate Director- Data Insights",
        "Data Analyst, Marketing", "Data Analyst (Hybrid/Remote)", "Principal Data Analyst (Remote)",
        "Director, Data Analyst - HYBRID", "Principal Data Analyst, AV Performance Analysis",
        "Principal Data Analyst", "ERM Data Analyst"
    ],
    "salary_year_avg": [650000.0, 336500.0, 255829.5, 232423.0, 217000.0, 205000.0, 189309.0, 189000.0, 186000.0, 184000.0]
}
df = pd.DataFrame(data)

# Sort the DataFrame based on 'salary_year_avg' to get the top paying roles
df_sorted = df.sort_values('salary_year_avg', ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(14, 8))
plt.barh(df_sorted['job_title'], df_sorted['salary_year_avg'], color='skyblue')
plt.xlabel('Average Yearly Salary ($)')
plt.ylabel('Job Title')
plt.title('Average Salary Distribution for Top 10 Paying Data Analyst Jobs in 2023')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest salary at the top
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

# Set the style to 'fivethirtyeight' for aesthetics
plt.style.use('fivethirtyeight')

# Increase the figure size for better readability
plt.figure(figsize=(12, 8))

df = pd.read_csv(r'C:\Users\colby\OneDrive\Documents\Data Analysis\SQL_Project_Data_Job_Analysis\data_viz\2_top_paying_roles.csv')
skills_list = df['skills'].str.split(',').tolist()


# Flatten the list of lists into a single list of skills
flat_skills_list = [skill for sublist in skills_list for skill in sublist]

# Convert the flat list into a Series and count the occurrences
skills_series = pd.Series(flat_skills_list)
skills_count = skills_series.value_counts()

# Apply the FiveThirtyEight style
style.use('fivethirtyeight')


import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'skills_count' is a Series with skills as the index and their counts as values
# Set the style to 'fivethirtyeight' for aesthetics
plt.style.use('fivethirtyeight')

# Create a horizontal bar chart
skills_count.head(10).plot(kind='barh', figsize=(10, 8), color='skyblue')

# Invert the y-axis to have the highest value at the top
plt.gca().invert_yaxis()

# Set the title and labels
plt.title('Top 10 Skills for Data Analyst Roles in 2023')
plt.xlabel('Frequency')
plt.ylabel('Skills')

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()

```
